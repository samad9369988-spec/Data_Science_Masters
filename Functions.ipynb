```python
l=[1,2,3,'sam','pw',[5,6,7,8]]
#create a function taking list as input and giving list as output with only numeric values
```

```python
def listnum(lst):
    n=[]
    for i in lst:
        if type(i)==list:
            for j in i:
                if type(j)==int or type(j)== float:
                    n.append(j)
        else :
            if type(i)==int or type(i)==float:
                n.append(i)
    return n 
```

```python
listnum(l)
```

# Generator functions

```python
range(1,10)
```

```python
for i in range(1,10):
    print(i)
```

```python
l=[1,2,3,4,'samad']
```

```python
def tst(a):
    n=[]
    for i in a:
        if type(i) ==int:
            n.append(i)
    return n
```

```python
tst(l)
```

```python
def tst(n):
    a,b=0,1
    for i in range(n):
        yield a 
        a,b=b,a+b
```

```python
tst(10)
```

```python
for i in tst(10):
    print(i)
```

```python
# The main advantage of gen fun is that it doesnot wait for the prep of complete data thereby blocking huge amount of memory rather it displays it one by one without storing the complete data
```

# Map Reduce and filter functions

```python
l=[1,2,3,4,5]
def sq(x):
    return x**2
list(map(sq,l))
```

```python
list(map(lambda x:x**2,l))
```

```python
l1=[1,2,3,4,5]
l2=[5,6,7,8,9,10]
```

```python
list(map(lambda x,y:x+y,l1,l2))
```

```python
s='pwskills'
list(map(lambda x:x.upper(),s))
```

```python
from functools import reduce
```
